```python
preds = [
    ('NeuroGSim', torch.load(f'../preds/{NAME}/inner_pred.pt', map_location='cpu')),
    ('Branch   ', torch.load(f'../preds/{NAME}/b_inner_pred.pt', map_location='cpu')),
    ('F2       ', torch.load(f'../preds/{NAME}/f2_inner_pred.pt', map_location='cpu')),
]
```

```python
inner_test_set, _ = torch.load(f'../data/{NAME}/protein_inner_plus_10K_4_0.5_25_4_f2_60_testing.pt', map_location='cpu')
queries, _, lb, ub = inner_test_set
sed = (lb+ub)/2
```

```python
print('=== RMSE / MAE / R2 ===')
print()
for label, pred in preds:
    print(f'{label}', end='\t')
    print(f'{metrics.rmse(lb, ub, pred):.3f}', end=' / ')
    print(f'{metrics.mae(lb, ub, pred):.3f}', end=' / ')
    print(f'{metrics.r2(lb, ub, pred):.5f}')
    print()
```

```python
print('=== rel. RMSE / rel. MAE ===')
print()
scale = sed+1
for label, pred in preds:
    print(f'{label}', end='\t')
    print(f'{metrics.rmse(lb/scale, ub/scale, pred/scale):.3f}', end=' / ')
    print(f'{metrics.mae(lb/scale, ub/scale, pred/scale):.3f}')
    print()
```

```python
err = metrics.err(lb, ub, preds[0][1])
plt.figure()
gsz = torch.tensor([g.num_nodes for g in queries])
viz.plot_corr_2d(sed, gsz, err, xmax1=50, xmax2=25, bin_size1=1, bin_size2=1, s=20, cmap='Reds', vmax=5, vmin=0)
plt.colorbar()
plt.xlabel('sed')
plt.ylabel('query size')
plt.title(f'Prediction Error (NeuroGSim)')
plt.show()
```
